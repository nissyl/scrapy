```python
import socks, socket
socks.setdefaultproxy(socks.PROXY_TYPE_SOCKS5, "127.0.0.1", 9150)
socket.socket = socks.socksocket
import requests
from bs4 import BeautifulSoup #htmlからタグを抽出するためのモジュール
import pandas as pd
import csv
import re
```

```python
# SE・インフラエンジニア・WEBエンジニアの検索ページの最後のページ番号を割り出す
url = 'https://doda.jp/DodaFront/View/JobSearchList/j_oc__03L/-preBtn__3/-page__1/'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
for parts in soup.findAll('p',{'class':'counter'}):
    parts = parts.find('span',{'class':'number'})
    page_num_se = -(-int(parts.string) // 50) #小数点は切り上げにする
    
```

```python
# 組込みソフトウェアの検索ページの最後のページ番号を割り出す
url = 'https://doda.jp/DodaFront/View/JobSearchList/j_oc__14L/-preBtn__3/-page__1/'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
for parts in soup.findAll('p',{'class':'counter'}):
    parts = parts.find('span',{'class':'number'})
    page_num_soft = -(-int(parts.string) // 50)
```

```python
# クリエイティブの検索ページの最後のページ番号を割り出す
url = 'https://doda.jp/DodaFront/View/JobSearchList/j_oc__08L/-preBtn__3/-page__1/'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
for parts in soup.findAll('p',{'class':'counter'}):
    parts = parts.find('span',{'class':'number'})
    page_num_creative = -(-int(parts.string) // 50)
```

```python
# それぞれページ数を確認
print(page_num_se)
print(page_num_soft)
print(page_num_creative)
```

```python
# SE・インフラエンジニア・WEBエンジニアの検索ページのURLを全て出す
search_url_se_list = []
page_num_se_sub = int(page_num_se) + 1 #1ページが足りなくなるので足す
for i in range(1,page_num_se_sub):
    num = str(i)
    url = "https://doda.jp/DodaFront/View/JobSearchList/j_oc__03L/-preBtn__3/-page__"+num+"/"
    search_url_se_list.append(url)
    print(url)

#出したURLをCSVで保存 search_url_se_list.csv
search_url_se_list = pd.DataFrame(search_url_se_list)
search_url_se_list.to_csv( 'search_url_se_list.csv',index=None,header=None) #列名を指定しないと1行目が0になる。
```

```python
# CSVで保存した検索ページ全てで企業URLを取得する
url_company_se_list = []
f = open('search_url_se_list.csv', 'r')#CSVファイルの読み込み
dataReader = csv.reader(f)#CSVファイルを読める形に置き換える
for row in dataReader:   #URLの入ったCSVを1行ずつ読み込んでくれる
    for url in row:   #CSVの1行に入っているURLを1行ずつ読み込む
        r = requests.get(url,timeout=60)#urlから、htmlの情報を取得する。
        soup = BeautifulSoup(r.content, 'html.parser')
        for box in soup.findAll('h2',{'class':'title clrFix'}):
            for companny_url in box.findAll('a',{'class':'_JobListToDetail'}):
                url_company = (re.sub("-tab__pr","-tab__jd",companny_url.get('href'))) #URLの終点部分をprからjdに変換する。欲しい情報は__jdにあるから
                url_company_se_list.append(url_company)
                print(url_company)

url_company_se_list = pd.DataFrame(url_company_se_list)
url_company_se_list.to_csv( 'url_company_se_list.csv',index=None,header=None) #列名を指定しないと1行目が0になる。                
```
